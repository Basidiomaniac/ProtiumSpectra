```python
from specdal import Collection, Spectrum, read
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# make sure we get high res plots
%config InlineBackend.figure_format = 'retina'
```

```python
import os

# read all the asd files into a Collections object
# white_reference00000 is from march 20
# whitereference00000 is from april 23 (which is all missing data anyway)
# other two folders have no white reference file
datadir = '/Users/niko/src/protium_spectra/asd/'
c = Collection(name = 'protium')
for f in [filename for filename in sorted(os.listdir(datadir)) if filename[-4:] == '.asd']:
    spectrum = Spectrum(filepath = os.path.join(datadir, f))
    c.append(spectrum)
```

```python
# take a peek at the data we imported
print(c['Protium09x00003'])
print(c['whitereference00000'])
```

```python
# check out the jumps in the spectra, let's correct these
c.plot(legend=False, xlim=(900, 1100), ylim=(0.2, 1.0))
plt.show()
```

```python
c.plot(legend=False, xlim=(1700, 1900), ylim=(0.2, 1.0))
plt.show()
```

```python
# clean up the data

# correct the jumps
c.jump_correct(splices = [1000, 1800], reference = 0)

c.plot(legend = False)
c.plot(legend=False, xlim=(900, 1100), ylim=(0.2, 1.0))
c.plot(legend=False, xlim=(1700, 1900), ylim=(0.2, 1.0))
plt.show()
```

```python
# flatten the index
nir = c.data
nir['wavelength'] = nir.index
nir.reset_index(drop = True, inplace = True)
```

```python
# where are the funky sample and replicate ids?
# a normal id is 'Protium' + 2-3 digits + 'x' + 5 digits
weirds = ~nir.columns.str.contains('^Protium[0-9]{2,3}x[0-9]{5}$')
nir.columns[weirds]
```

```python
# check if there's any other weirdness besides the duplicate measurements
non_dups = ~nir.columns[weirds].str.contains('(1)')
nir.columns[weirds][non_dups]
```

```python
# looks like it's just the white references and these duplicate measurements
# just drop the weirds for now
wv = nir['wavelength']
nir = nir.iloc[:, ~weirds]
nir['wavelength'] = wv
nir
```

```python
# can we cast wavelength to int?
non_int = nir['wavelength'] % 1 > 0
sum(non_int) # yes!
```

```python
nir['wavelength'] = pd.to_numeric(nir['wavelength'], downcast = 'integer')
```

```python
# pivot longer so each observation is a row
nir_long = pd.melt(nir, id_vars = ['wavelength'], var_name = 'sample_raw', value_name = 'reflectance')

nir_long
```

```python
# separate out sample and replicate ids
nir_long[['sample_id', 'replicate']] = nir_long['sample_raw'].str.split('x', expand = True)
nir_long['sample_id'] = nir_long['sample_id'].str.split('Protium', expand = True)[1]
nir_long['sample_id'] = pd.to_numeric(nir_long['sample_id'])
nir_long['replicate'] = pd.to_numeric(nir_long['replicate'])
nir_long = nir_long.drop('sample_raw', axis = 1)
nir_long = nir_long[['sample_id', 'replicate', 'wavelength', 'reflectance']]
print(nir_long.dtypes)
nir_long
```

```python
# make a wide version of this dataframe for LDA downstream
nir_wide = nir_long.pivot(index = ['sample_id', 'replicate'], columns = 'wavelength')
nir_wide
```

```python
# flatten all the multilevel indices and start wavelength columns with a string... 'nm' is nanometers
nir_wide.reset_index(inplace = True, col_level = 1)
nir_wide.columns = nir_wide.columns.droplevel()
nir_wide.columns = ['nm' + str(c) if isinstance(c, int) else c for c in nir_wide.columns]
nir_wide
```

```python
# how do the reflectance readings look?
nir_na = nir_wide[nir_wide.isna().any(axis = 1)]
nir_na
```

```python
nir_inf = nir_wide[np.isinf(nir_wide).any(axis = 1)]
nir_inf
```

```python
inf_samples = nir_inf['sample_id'].unique()
print(np.sort(inf_samples))
print(len(inf_samples))
```

```python
# looks like samples 100-196 are broken, drop these for now
nir_long = nir_long.loc[nir_long.sample_id.isin(inf_samples) == False,]
nir_long
```

```python
nir_wide = nir_wide.loc[nir_wide.sample_id.isin(inf_samples) == False,]
nir_wide
```

```python
# crop to appetizer and mystery for now
# let's call sample 439 the mystery one
appetizer_samples = list(range(1, 52)) + [439]
nir_long_app = nir_long.loc[nir_long['sample_id'].isin(appetizer_samples),:]
nir_long_app
```

```python
nir_wide_app = nir_wide.loc[nir_wide['sample_id'].isin(appetizer_samples),:]
nir_wide_app
```

```python
# write long and wide versions to csv
nir_long.to_csv('data/protium_clean_reflectance_long.csv', index = False)
nir_wide.to_csv('data/protium_clean_reflectance_wide.csv', index = False)
nir_long_app.to_csv('data/protium_clean_reflectance_long_appetizer.csv', index = False)
nir_wide_app.to_csv('data/protium_clean_reflectance_wide_appetizer.csv', index = False)
```
